```python
def cleaning_pipeline(data: pl.DataFrame, output_path: str):
    """Data cleaning and processing pipeline

    Args:
        data (pl.DataFrame): input dataset that needs to be pre-processed
        output_path (str): location of where to save the parquet file
    """

    
    # Drop unnecessary columns
    columns_to_drop = [
        'Asymmetrique Activity Score', 'Asymmetrique Profile Score',
        'Asymmetrique Activity Index', 'Asymmetrique Profile Index'
    ]
    data = data.drop(columns_to_drop)

    # Identify categorical columns
    categorical_cols = [col for col in data.columns if data[col].dtype in [pl.Utf8, pl.Categorical]]
    
    # Handle "Select" values in categorical columns
    data = data.with_columns([
        pl.when(pl.col(col) == "Select").then(None).otherwise(pl.col(col)).alias(col)
        for col in categorical_cols
    ])

    # Fill nulls for multiple columns
    data = data.with_columns([pl.col("TotalVisits").fill_null(data["TotalVisits"].mean()).alias("TotalVisits"),
                              pl.col("Page Views Per Visit").fill_null(data["Page Views Per Visit"].mean()).alias("Page Views Per Visit"),
                              pl.col("Country").fill_null("India").alias("Country"),
                              pl.col("City").fill_null("Mumbai").alias("City"),
                              pl.col("Lead Source").fill_null("Reference").alias("Lead Source"),
                              pl.col("Last Activity").fill_null(data["Last Activity"].mode()[0]).alias("Last Activity"),
                              pl.col("Specialization").fill_null("Unknown").alias("Specialization"),
                              pl.col("How did you hear about X Education").fill_null("Unknown").alias("How did you hear about X Education"),
                              pl.col("What is your current occupation").fill_null(data["What is your current occupation"].mode()[0]).alias("What is your current occupation"),
                              pl.col("What matters most to you in choosing a course").fill_null(
                            data["What matters most to you in choosing a course"].mode()[0]
                            ).alias("What matters most to you in choosing a course"),
                              pl.col("Tags").fill_null("None").alias("Tags"),
                              pl.col("Lead Quality").fill_null("None").alias("Lead Quality"),
                              pl.col("Lead Profile").fill_null("None").alias("Lead Profile")
        ])

    data.write_parquet(output_path)
```

```python
output = "../data/supervised_clean_data.parquet"
cleaning_pipeline(data, output)
```
